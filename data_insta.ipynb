```python
import requests
import json

comment_url = "https://instagram-scraper-api2.p.rapidapi.com/v1/comments"

headers = {
    "x-rapidapi-key": "a8c4ed5b7dmsh9810575efeb725fp186d0ajsn0875bf677f8b",
    "x-rapidapi-host": "instagram-scraper-api2.p.rapidapi.com"
}

comment_data = []
with open(f'{hastag}.csv', 'w') as f:
    f.write("Username;Text\n")
    for code in codes:
        querystring = {"code_or_id_or_url": f"{code}"}
        response_comment = requests.get(comment_url, headers=headers, params=querystring)

        if response_comment.status_code == 200:
            data = response_comment.json()
            print("raw data \n", json.dumps(data, indent=2))

            if 'data' in data:
                additional_data = data["data"].get("additional_data", {})
                comments = data["data"].get("items", [])

                caption = additional_data.get('caption', {})
                hashtags = caption.get('hashtags', [])
                text = caption.get('text', '')
                user = caption.get('user', {})
                username = user.get('username', '')

                if hashtags and text and username:
                    print(f"hashtags : {hashtags}")
                    print(f"text: {text}")
                    print(f"username: {username}")
                    f.write(f"{username};{text}\n")
                    comment_data.append({
                        "username": username,
                        "text": text,
                        "hashtags": hashtags
                    })

                for comment in comments:
                    comment_username = comment["user"]["username"]
                    comment_text = comment["text"]
                    print("Username:", comment_username)
                    print("Text:", comment_text)
                    print("----")
                    f.write(f"{comment_username};{comment_text}\n")
                    comment_data.append(comment)
        else:
            print(f"Failed to retrieve comments for code {code}. HTTP Status Code: {response_comment.status_code}")
```
